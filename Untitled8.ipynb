```python
from statsmodels.tsa.stattools import adfuller
```

```python
test_result=adfuller(data1['price'])
```

```python
def adfuller_test(price):
    results=adfuller(price)
    labels=['ADF test statistic','p-value','#Lags used','Number of observation used']
    for value,label in zip(results,labels):
        print(label+' : '+str(value))
    if results[1] <=0.05:
        print("strong evidence that it is  stationary")
    else:
        print("weak evidence that it is stationary")
```

```python
adfuller_test(data1['price'])
```
